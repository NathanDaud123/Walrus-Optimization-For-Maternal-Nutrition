```python
import os
import pandas as pd

# Assuming the DataFrames mp, sn, sh, sy, plk are already defined as shown in the initial query

def write_menu_to_csv(best, filename='recommended_menu.csv'):
    # Create an empty list to store menu information
    menu_data = []
    
    # Define meal times for clarity
    meal_times = ['Makan Pagi', 'Makan Siang', 'Makan Malam']
    
    # Loop through each meal time
    for i, meal_time in enumerate(meal_times):
        start_index = i * 5
        # Extract the menu items for the current meal
        meal_items = [
            ('Makanan pokok', mp, best[start_index]),
            ('Sumber Nabati', sn, best[start_index + 1]),
            ('Sumber Hewani', sh, best[start_index + 2]),
            ('Sayuran', sy, best[start_index + 3]),
            ('Pelengkap', plk, best[start_index + 4]),
        ]
        
        # Loop through each menu item and append its details to the menu_data list
        for category, df, index in meal_items:
            menu_data.append({
                'Waktu': meal_time,
                'Kategori': category,
                'Menu': df['menu'][round(index)-1],
                'Karbo': df['carbo'][round(index)-1],
                'Energi': df['energy'][round(index)-1],
                'Lemak': df['fat'][round(index)-1],
                'Protein': df['protein'][round(index)-1],
                'Harga': df['price'][round(index)-1],
            })
    
    # Convert the menu_data list to a DataFrame
    menu_df = pd.DataFrame(menu_data)
    
    # Check if the file exists to determine if headers should be written
    file_exists = os.path.isfile(filename)
    
    # Write the DataFrame to a CSV file in append mode
    menu_df.to_csv(filename, mode='a', index=False, header=not file_exists, encoding='utf-8')
    
# Example usage
# best_individual = [1, 2, 3, 4, 5, 6, 7, 8, 9, 1, 2, 3, 4, 5, 6]  # This should be replaced with the actual best individual
# write_menu_to_csv(best_individual)
```

### Walrus Optimization Algorithm (WaOA)

```python
import numpy as np

# Define the objective function (to be minimized)
# def objective_function(x):
#     # Placeholder for the actual objective function
#     # Replace this with the actual function to be optimized
#     return np.sum(x**2)

# Initialize the population
def initialize_population(pop_size, dim, lower_bound, upper_bound):
    return lower_bound + (upper_bound - lower_bound) * np.random.rand(pop_size, dim)

def calculate_beta(t, T):
    exponent = -((t - (T / 2)) / T) * 10
    return 1 - (1 / (1 + np.exp(exponent)))

# Calculate fitness for the population
def calculate_fitness(population):
    return np.array([cost_function(individual) for individual in population])

# Update position during exploration phase
def exploration_update(position, alpha, beta, r3, X_best, X_worst):
    return position + (X_best - X_worst) * beta * r3**2

# Update position during exploitation phase (roosting behavior)
def exploitation_update(position, alpha, male, X_best):
    return position + alpha * (male - position) + (1 - alpha) * (X_best - position)

def roosting_update(female, male, alpha, X_best):
    # Female walrus is influenced by male walrus and the lead walrus (X_best)
    # As the process of iteration, the female walrus is gradually influenced less by the mate and more by the leader.
    return female + alpha * (male - female) + (1 - alpha) * (X_best - female)

def gathering_update(X, X_best, X_second_best, a, b):
    # Walruses cooperate to forage and move according to the location of other walruses in the population
    # Sharing location information can help the whole walrus herd to find the sea area with higher food abundance.
    X1 = X_best - a * b * (X_best - X)
    X2 = X_second_best - a * b * (X_second_best - X)
    return (X1 + X2) / 2

def fleeing_update(X, X_best, r4):
    # Walruses flee from their current activity area based on danger signals from their peers.
    # This behavior occurs in the late iteration of the WO, and a certain degree of perturbation to the population helps walruses to conduct global exploration.
    return X * r4 - abs(X_best - X) * r4**2


# Main loop of the Walrus Optimization Algorithm
def walrus_optimization_algorithm(pop_size, dim, max_iter, lower_bound, upper_bound, cost_function):
    # Initialize population
    population = initialize_population(pop_size, dim, lower_bound, upper_bound)
    fitness = calculate_fitness(population)
    
    # Initialize best solution
    best_solution = population[np.argmin(fitness)]
    best_fitness = np.min(fitness)
    
    # Main optimization loop
    for t in range(max_iter):
        # Calculate danger signal and safety signal
        alpha = 1 - t / max_iter
        beta = calculate_beta(t, max_iter)
        A = 2 * alpha
        R = 2 * np.random.rand() - 1
        danger_signal = A * R
        safety_signal = np.random.rand()
        
        # Exploration phase
        # Migration
        if abs(danger_signal) >= 1:
            for i in range(pop_size):
                r3 = np.random.rand()
                vig1, vig2 = np.random.choice(population, 2, replace=False)
                population[i] = exploration_update(population[i], alpha, beta, r3, vig1, vig2)

        # Exploitation phase
        else:
            if safety_signal >= 0.5:
                # Breeding behavior
                for i in range(pop_size // 2):  # Assuming half the population is male
                    male = population[i]
                    female = population[pop_size // 2 + i]
                    population[i] = roosting_update(male, female, alpha, best_solution)
            else:
                # Foraging behavior
                if abs(danger_signal) >= 0.5:
                    # Fleeing behavior
                    for i in range(pop_size):
                        r4 = np.random.rand()
                        population[i] = fleeing_update(population[i], best_solution, r4)
                else:
                    # Gathering behavior
                    for i in range(pop_size):
                        a = beta * np.random.rand() - beta
                        b = np.tan(np.pi * np.random.rand())
                        X_second_best = population[np.argsort(fitness)[-2]]  # Assuming second best is the one before the last in a sorted list
                        population[i] = gathering_update(population[i], best_solution, X_second_best, a, b)
        
        # Evaluate the fitness of the new solutions
        fitness = calculate_fitness(population)
        
        # Update the best solution found so far
        current_best = population[np.argmin(fitness)]
        current_best_fitness = np.min(fitness)
        if current_best_fitness < best_fitness:
            best_solution = current_best
            best_fitness = current_best_fitness
    
    return best_solution, best_fitness

# Algorithm parameters
pop_size = 50  # Population size
dim = 30  # Dimension of the problem
max_iter = 100  # Maximum number of iterations
lower_bound = -100  # Lower bound of decision variables
upper_bound = 100  # Upper bound of decision variables

# Run the algorithm
best_solution, best_fitness = walrus_optimization_algorithm(pop_size, dim, max_iter, lower_bound, upper_bound)
print("Best solution:", best_solution)
print("Best fitness:", best_fitness)
```

```python
import numpy as np

def WOA(SearchAgents, Max_iterations, lowerbound, upperbound, dimension, fitness):
    lowerbound = np.ones(dimension) * lowerbound
    upperbound = np.ones(dimension) * upperbound

    # INITIALIZATION
    X = np.random.rand(SearchAgents, dimension) * (upperbound - lowerbound) + lowerbound
    fit = np.apply_along_axis(fitness, 1, X)

    fbest = np.inf
    Xbest = None

    fitness_best_so_far = []
    sol_best_so_far = []
    average = []

    for t in range(1, Max_iterations + 1):
        # update the best candidate solution
        best = np.min(fit)
        location = np.argmin(fit)
        if t == 1 or best < fbest:
            fbest = best
            Xbest = X[location, :]

        # print("ini awal program\t" + str(fbest))
        
        SW = Xbest  # strongest walrus with best value for objective function

        for i in range(SearchAgents):
            # PHASE 1: FEEDING STRATEGY (EXPLORATION)
            I = np.round(1 + np.random.rand())
            X_P1 = X[i, :] + np.random.rand(dimension) * (SW - I * X[i, :])
            X_P1 = np.clip(X_P1, lowerbound, upperbound)

            F_P1 = fitness(X_P1)
            if F_P1 < fit[i]:
                X[i, :] = X_P1
                fit[i] = F_P1

            # PHASE 2: MIGRATION
            I = np.round(1 + np.random.rand())
            K = np.random.permutation(SearchAgents)
            K = K[K != i]
            X_K = X[K[0], :]
            F_RAND = fit[K[0]]
            if fit[i] > F_RAND:
                X_P2 = X[i, :] + np.random.rand() * (X_K - I * X[i, :])
            else:
                X_P2 = X[i, :] + np.random.rand() * (X[i, :] - X_K)

            X_P2 = np.clip(X_P2, lowerbound, upperbound)
            F_P2 = fitness(X_P2)
            if F_P2 < fit[i]:
                X[i, :] = X_P2
                fit[i] = F_P2

            # PHASE 3: ESCAPING AND FIGHTING AGAINST PREDATORS (EXPLOITATION)
            LO_LOCAL = lowerbound / t
            HI_LOCAL = upperbound / t
            I = np.round(1 + np.random.rand())

            X_P3 = X[i, :] + LO_LOCAL + np.random.rand() * (HI_LOCAL - LO_LOCAL)
            X_P3 = np.clip(X_P3, LO_LOCAL, HI_LOCAL)
            X_P3 = np.clip(X_P3, lowerbound, upperbound)

            F_P3 = fitness(X_P3)
            if F_P3 < fit[i]:
                X[i, :] = X_P3
                fit[i] = F_P3

        best = np.min(fit)
        location = np.argmin(fit)
        if t == 1 or best < fbest:
            fbest = best
            Xbest = X[location, :]

        # print("ini akhir program\t" + str(fbest))

        sol_best_so_far.append(Xbest)
        fitness_best_so_far.append(fbest)
    
        unique_values, first_occurrence_indexes = np.unique(fitness_best_so_far, return_index=True)
        last_three_unique_values = unique_values[:3]
        last_three_indexes = first_occurrence_indexes[:3]
        
    print(unique_values)
    print(last_three_unique_values)
    print(last_three_indexes)
    print("======================")
    print(fitness_best_so_far)

    # print(sol_best_so_far)
    for individual in last_three_indexes:
        print("Indeks dari 3 Nilai Unik Terakhir:", individual)
        print("="*100)
        print("Cost Unik Terakhir:", fitness_best_so_far[individual])
        daftar_makanan(sol_best_so_far[individual])
        write_menu_to_csv(sol_best_so_far[individual])
    
    Best_score = fbest
    Best_pos = Xbest
    WOA_curve = fitness_best_so_far

    return Best_score, Best_pos, WOA_curve
```
